```python
import torch
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.auto import tqdm
from transformers import (
    T5Tokenizer, 
    T5ForConditionalGeneration, 
    AdamW, 
    get_linear_schedule_with_warmup
)
from torch.utils.data import Dataset, DataLoader
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import corpus_bleu, sentence_bleu
from nltk.tokenize import word_tokenize
import os
import logging
import time
import re
from datetime import datetime

# Set up logging
logging.basicConfig(
    level=logging.INFO,
    format='%(asctime)s - %(name)s - %(levelname)s - %(message)s',
    handlers=[
        logging.FileHandler(f"t5_summarization_{datetime.now().strftime('%Y%m%d_%H%M%S')}.log"),
        logging.StreamHandler()
    ]
)
logger = logging.getLogger(__name__)

# Set random seeds for reproducibility
def set_seed(seed=42):
    """Set all seeds to make results reproducible"""
    torch.manual_seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed_all(seed)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False
    np.random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)

set_seed(42)

# Check for GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
logger.info(f"Using device: {device}")

# Configuration
class Config:
    """Configuration class to store the training parameters"""
    model_name = "t5-base"
    input_max_len = 512
    output_max_len = 150
    train_batch_size = 4
    valid_batch_size = 4
    test_batch_size = 4
    gradient_accumulation_steps = 2
    num_epochs = 5
    learning_rate = 3e-4
    weight_decay = 0.01
    warmup_ratio = 0.1
    max_grad_norm = 1.0
    eval_every = 500
    save_every = 500
    early_stopping_patience = 3
    output_dir = "t5_summarization_model"

# Dataset class
class NewsSummaryDataset(Dataset):
    """Dataset for news summarization"""
    def __init__(self, dataframe, tokenizer, config, is_test=False):
        self.tokenizer = tokenizer
        self.data = dataframe
        self.text = dataframe.paragraphs
        self.summary = dataframe.summary
        self.config = config
        self.is_test = is_test

    def __len__(self):
        return len(self.data)

    def __getitem__(self, index):
        text = self.text[index]
        summary = self.summary[index]

        # Add prefix to the input for T5
        text_with_prefix = "summarize: " + text

        # Tokenize inputs
        text_encoding = self.tokenizer(
            text_with_prefix, 
            max_length=self.config.input_max_len, 
            padding="max_length", 
            truncation=True, 
            return_attention_mask=True, 
            return_tensors="pt"
        )

        # Tokenize targets
        target_encoding = self.tokenizer(
            summary, 
            max_length=self.config.output_max_len, 
            padding="max_length", 
            truncation=True, 
            return_attention_mask=True, 
            return_tensors="pt"
        )

        input_ids = text_encoding["input_ids"].squeeze()
        attention_mask = text_encoding["attention_mask"].squeeze()
        
        target_ids = target_encoding["input_ids"].squeeze()
        # Replace padding token id's with -100 for the loss calculation
        target_ids[target_ids == self.tokenizer.pad_token_id] = -100

        return {
            "input_ids": input_ids,
            "attention_mask": attention_mask,
            "labels": target_ids,
            "original_text": text,
            "original_summary": summary
        }

# Data preprocessing function
def preprocess_text(text):
    """Preprocess text data"""
    if not isinstance(text, str):
        return ""
    
    # Convert to lowercase
    text = text.lower()
    
    # Replace newlines and tabs with spaces
    text = text.replace('\n', ' ')
    text = text.replace('\t', ' ')
    
    # Remove URLs
    text = re.sub(r'https?://\S+|www\.\S+', '', text)
    
    # Remove HTML tags
    text = re.sub(r'<.*?>', '', text)
    
    # Normalize whitespace
    text = ' '.join(text.split())
    
    return text

# Load and prepare data
def load_and_prepare_data(train_path, dev_path, test_path):
    """Load and prepare the datasets"""
    # Load data
    logger.info("Loading datasets...")
    train_df = pd.read_csv(train_path)
    dev_df = pd.read_csv(dev_path)
    test_df = pd.read_csv(test_path)
    
    # Verify data
    logger.info(f"Train data: {train_df.shape}, Dev data: {dev_df.shape}, Test data: {test_df.shape}")
    
    # Check for null values
    for df, name in zip([train_df, dev_df, test_df], ['Train', 'Dev', 'Test']):
        null_counts = df.isnull().sum()
        logger.info(f"{name} null counts: {null_counts}")
        # Fill null values
        df.fillna("", inplace=True)
    
    # Convert to string
    for df in [train_df, dev_df, test_df]:
        df["paragraphs"] = df["paragraphs"].astype(str)
        df["summary"] = df["summary"].astype(str)
    
    # Preprocess text
    logger.info("Preprocessing data...")
    for df in [train_df, dev_df, test_df]:
        df["paragraphs"] = df["paragraphs"].apply(preprocess_text)
        df["summary"] = df["summary"].apply(preprocess_text)
    
    # Data statistics
    for df, name in zip([train_df, dev_df, test_df], ['Train', 'Dev', 'Test']):
        input_lens = df['paragraphs'].apply(len)
        target_lens = df['summary'].apply(len)
        logger.info(f"{name} input length stats: mean={input_lens.mean():.1f}, min={input_lens.min()}, max={input_lens.max()}")
        logger.info(f"{name} summary length stats: mean={target_lens.mean():.1f}, min={target_lens.min()}, max={target_lens.max()}")
    
    return train_df, dev_df, test_df

# Create dataloaders
def create_dataloaders(train_df, dev_df, test_df, tokenizer, config):
    """Create DataLoader objects for training, validation and testing"""
    logger.info("Creating datasets and dataloaders...")
    
    # Create datasets
    train_dataset = NewsSummaryDataset(train_df, tokenizer, config)
    dev_dataset = NewsSummaryDataset(dev_df, tokenizer, config)
    test_dataset = NewsSummaryDataset(test_df, tokenizer, config)
    
    # Create dataloaders
    train_loader = DataLoader(
        train_dataset, 
        batch_size=config.train_batch_size, 
        shuffle=True,
        num_workers=2 if torch.cuda.is_available() else 0,
        pin_memory=True if torch.cuda.is_available() else False
    )
    
    dev_loader = DataLoader(
        dev_dataset, 
        batch_size=config.valid_batch_size,
        num_workers=2 if torch.cuda.is_available() else 0,
        pin_memory=True if torch.cuda.is_available() else False
    )
    
    test_loader = DataLoader(
        test_dataset, 
        batch_size=config.test_batch_size,
        num_workers=2 if torch.cuda.is_available() else 0,
        pin_memory=True if torch.cuda.is_available() else False
    )
    
    return train_loader, dev_loader, test_loader

# Training function
def train_epoch(model, dataloader, optimizer, scheduler, config, epoch):
    """Train the model for one epoch"""
    model.train()
    total_loss = 0
    epoch_start_time = time.time()
    
    progress_bar = tqdm(dataloader, desc=f"Epoch {epoch+1} [Training]")
    
    for batch_idx, batch in enumerate(progress_bar):
        input_ids = batch["input_ids"].to(device)
        attention_mask = batch["attention_mask"].to(device)
        labels = batch["labels"].to(device)
        
        # Forward pass
        outputs = model(
            input_ids=input_ids,
            attention_mask=attention_mask,
            labels=labels
        )
        loss = outputs.loss / config.gradient_accumulation_steps
        loss.backward()
        
        if (batch_idx + 1) % config.gradient_accumulation_steps == 0:
            torch.nn.utils.clip_grad_norm_(model.parameters(), config.max_grad_norm)
            optimizer.step()
            scheduler.step()
            optimizer.zero_grad()
        
        # Update statistics
        total_loss += loss.item() * config.gradient_accumulation_steps
        
        # Update progress bar
        progress_bar.set_postfix({
            'loss': f"{loss.item() * config.gradient_accumulation_steps:.4f}",
            'avg_loss': f"{total_loss / (batch_idx + 1):.4f}"
        })
    
    epoch_time = time.time() - epoch_start_time
    logger.info(f"Epoch {epoch+1} completed in {epoch_time:.2f} seconds. Avg. Loss: {total_loss / len(dataloader):.4f}")
    
    return total_loss / len(dataloader)

# Evaluation function
def evaluate(model, dataloader, config, tokenizer=None, compute_metrics=False):
    """Evaluate the model on the given dataloader"""
    model.eval()
    total_loss = 0
    
    all_predictions = []
    all_references = []
    
    progress_bar = tqdm(dataloader, desc="Evaluation")
    
    with torch.no_grad():
        for batch in progress_bar:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            labels = batch["labels"].to(device)
            
            outputs = model(
                input_ids=input_ids,
                attention_mask=attention_mask,
                labels=labels
            )
            
            loss = outputs.loss
            total_loss += loss.item()
            
            progress_bar.set_postfix({'loss': f"{loss.item():.4f}"})
            
            # Generate summaries for metric calculation
            if compute_metrics and tokenizer is not None:
                generated_ids = model.generate(
                    input_ids=input_ids,
                    attention_mask=attention_mask,
                    max_length=config.output_max_len,
                    num_beams=4,
                    repetition_penalty=2.5,
                    length_penalty=1.0,
                    early_stopping=True
                )
                
                decoded_preds = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=True) for g in generated_ids]
                decoded_labels = [tokenizer.decode(l[l != -100], skip_special_tokens=True, clean_up_tokenization_spaces=True) for l in labels]
                
                all_predictions.extend(decoded_preds)
                all_references.extend(decoded_labels)
    
    # Compute metrics
    metrics = {}
    if compute_metrics and all_predictions:
        metrics = compute_evaluation_metrics(all_predictions, all_references)
    
    return total_loss / len(dataloader), metrics, all_predictions, all_references

# Compute evaluation metrics
def compute_evaluation_metrics(predictions, references):
    """Compute ROUGE and BLEU scores"""
    # Initialize ROUGE scorer
    rouge_scorer_instance = rouge_scorer.RougeScorer(['rouge1', 'rouge2', 'rougeL'], use_stemmer=True)
    rouge_scores = [rouge_scorer_instance.score(ref, pred) for ref, pred in zip(references, predictions)]
    
    # Calculate average ROUGE scores
    avg_rouge_scores = {
        "rouge1": {
            "precision": np.mean([score['rouge1'].precision for score in rouge_scores]),
            "recall": np.mean([score['rouge1'].recall for score in rouge_scores]),
            "fmeasure": np.mean([score['rouge1'].fmeasure for score in rouge_scores])
        },
        "rouge2": {
            "precision": np.mean([score['rouge2'].precision for score in rouge_scores]),
            "recall": np.mean([score['rouge2'].recall for score in rouge_scores]),
            "fmeasure": np.mean([score['rouge2'].fmeasure for score in rouge_scores])
        },
        "rougeL": {
            "precision": np.mean([score['rougeL'].precision for score in rouge_scores]),
            "recall": np.mean([score['rougeL'].recall for score in rouge_scores]),
            "fmeasure": np.mean([score['rougeL'].fmeasure for score in rouge_scores])
        }
    }
    
    # Calculate BLEU score
    tokenized_references = [[word_tokenize(ref.lower())] for ref in references]
    tokenized_predictions = [word_tokenize(pred.lower()) for pred in predictions]
    
    try:
        bleu_score = corpus_bleu(tokenized_references, tokenized_predictions)
    except Exception as e:
        logger.error(f"Error calculating BLEU score: {e}")
        bleu_score = 0.0
    
    # Combine metrics
    metrics = {
        "rouge1_precision": avg_rouge_scores["rouge1"]["precision"],
        "rouge1_recall": avg_rouge_scores["rouge1"]["recall"],
        "rouge1_f1": avg_rouge_scores["rouge1"]["fmeasure"],
        "rouge2_precision": avg_rouge_scores["rouge2"]["precision"],
        "rouge2_recall": avg_rouge_scores["rouge2"]["recall"],
        "rouge2_f1": avg_rouge_scores["rouge2"]["fmeasure"],
        "rougeL_precision": avg_rouge_scores["rougeL"]["precision"],
        "rougeL_recall": avg_rouge_scores["rougeL"]["recall"],
        "rougeL_f1": avg_rouge_scores["rougeL"]["fmeasure"],
        "bleu": bleu_score
    }
    
    return metrics

# Save checkpoint
def save_checkpoint(model, tokenizer, optimizer, scheduler, epoch, val_loss, metrics, config, filename):
    """Save model checkpoint"""
    if not os.path.exists(config.output_dir):
        os.makedirs(config.output_dir)
    
    logger.info(f"Saving model checkpoint to {filename}")
    
    checkpoint = {
        'epoch': epoch,
        'model_state_dict': model.state_dict(),
        'optimizer_state_dict': optimizer.state_dict(),
        'scheduler_state_dict': scheduler.state_dict() if scheduler else None,
        'val_loss': val_loss,
        'metrics': metrics,
        'config': vars(config) if hasattr(config, '__dict__') else config
    }
    
    torch.save(checkpoint, filename)
    tokenizer.save_pretrained(config.output_dir)

# Load checkpoint
def load_checkpoint(model, optimizer, scheduler, filename):
    """Load model checkpoint"""
    logger.info(f"Loading model checkpoint from {filename}")
    
    checkpoint = torch.load(filename, map_location=device)
    
    model.load_state_dict(checkpoint['model_state_dict'])
    optimizer.load_state_dict(checkpoint['optimizer_state_dict'])
    
    if checkpoint['scheduler_state_dict'] and scheduler:
        scheduler.load_state_dict(checkpoint['scheduler_state_dict'])
    
    epoch = checkpoint['epoch']
    val_loss = checkpoint['val_loss']
    metrics = checkpoint.get('metrics', {})
    
    return model, optimizer, scheduler, epoch, val_loss, metrics

# Visualize training history
def visualize_training_history(history):
    """Visualize training history"""
    plt.figure(figsize=(15, 7))
    
    # Plot Loss
    plt.subplot(1, 2, 1)
    epochs = range(1, len(history['train_loss']) + 1)
    plt.plot(epochs, history['train_loss'], 'b-o', label='Train Loss')
    plt.plot(epochs, history['val_loss'], 'r-o', label='Validation Loss')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Loss', fontsize=12)
    plt.title('Training and Validation Loss', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    # Plot Metrics
    plt.subplot(1, 2, 2)
    plt.plot(epochs, history['rouge1_f1'], 'g-o', label='ROUGE-1 F1')
    plt.plot(epochs, history['rouge2_f1'], 'm-o', label='ROUGE-2 F1')
    plt.plot(epochs, history['rougeL_f1'], 'c-o', label='ROUGE-L F1')
    plt.plot(epochs, history['bleu'], 'y-o', label='BLEU')
    plt.xlabel('Epoch', fontsize=12)
    plt.ylabel('Score', fontsize=12)
    plt.title('Evaluation Metrics', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    
    plt.tight_layout()
    plt.savefig('training_history.png', dpi=300, bbox_inches='tight')
    logger.info("Training history visualization saved to 'training_history.png'")
    
    try:
        from IPython import get_ipython
        if get_ipython() is not None:
            plt.show()
    except:
        pass
    
    plt.close()

# Analyze examples
def analyze_examples(model, dataloader, tokenizer, config, num_examples=5):
    """Analyze a few examples from the dataloader"""
    model.eval()
    examples = []
    
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch["input_ids"].to(device)
            attention_mask = batch["attention_mask"].to(device)
            
            # Generate summaries
            generated_ids = model.generate(
                input_ids=input_ids,
                attention_mask=attention_mask,
                max_length=config.output_max_len,
                num_beams=4,
                repetition_penalty=2.5,
                length_penalty=1.0,
                early_stopping=True
            )
            
            # Decode generated summaries and reference summaries
            original_texts = batch["original_text"]
            original_summaries = batch["original_summary"]
            generated_summaries = [tokenizer.decode(g, skip_special_tokens=True, clean_up_tokenization_spaces=True) for g in generated_ids]
            
            # Add examples
            for orig_text, orig_summary, gen_summary in zip(original_texts, original_summaries, generated_summaries):
                examples.append({
                    "original_text": orig_text,
                    "original_summary": orig_summary,
                    "generated_summary": gen_summary
                })
                
                if len(examples) >= num_examples:
                    return examples
    
    return examples

# Main training function
def train_and_evaluate(config):
    """Main function to train and evaluate the model"""
    # Load and prepare data
    train_df, dev_df, test_df = load_and_prepare_data('train.csv', 'dev.csv', 'test.csv')
    
    # Initialize tokenizer and model
    logger.info(f"Initializing tokenizer and model from {config.model_name}...")
    tokenizer = T5Tokenizer.from_pretrained(config.model_name)
    model = T5ForConditionalGeneration.from_pretrained(config.model_name)
    model.to(device)
    
    # Create dataloaders
    train_loader, dev_loader, test_loader = create_dataloaders(train_df, dev_df, test_df, tokenizer, config)
    
    # Initialize optimizer and scheduler
    optimizer = AdamW(
        model.parameters(),
        lr=config.learning_rate,
        weight_decay=config.weight_decay
    )
    
    # Calculate total training steps
    total_steps = len(train_loader) * config.num_epochs // config.gradient_accumulation_steps
    warmup_steps = int(total_steps * config.warmup_ratio)
    
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=warmup_steps,
        num_training_steps=total_steps
    )
    
    # Create output directory if it doesn't exist
    if not os.path.exists(config.output_dir):
        os.makedirs(config.output_dir)
    
    # Training loop
    logger.info("Starting training...")
    
    history = {
        'train_loss': [],
        'val_loss': [],
        'rouge1_f1': [],
        'rouge2_f1': [],
        'rougeL_f1': [],
        'bleu': []
    }
    
    best_val_loss = float('inf')
    early_stopping_counter = 0
    
    try:
        for epoch in range(config.num_epochs):
            # Train
            train_loss = train_epoch(model, train_loader, optimizer, scheduler, config, epoch)
            history['train_loss'].append(train_loss)
            
            # Evaluate
            val_loss, val_metrics, _, _ = evaluate(model, dev_loader, config, tokenizer, compute_metrics=True)
            history['val_loss'].append(val_loss)
            
            # Update history with metrics
            for metric_name in ['rouge1_f1', 'rouge2_f1', 'rougeL_f1', 'bleu']:
                history[metric_name].append(val_metrics.get(metric_name, 0.0))
            
            # Log metrics
            log_str = f"Epoch {epoch+1}/{config.num_epochs} - "
            log_str += f"Train Loss: {train_loss:.4f}, Val Loss: {val_loss:.4f}, "
            log_str += f"ROUGE-1 F1: {val_metrics.get('rouge1_f1', 0.0):.4f}, "
            log_str += f"ROUGE-2 F1: {val_metrics.get('rouge2_f1', 0.0):.4f}, "
            log_str += f"ROUGE-L F1: {val_metrics.get('rougeL_f1', 0.0):.4f}, "
            log_str += f"BLEU: {val_metrics.get('bleu', 0.0):.4f}"
            logger.info(log_str)
            
            # Save checkpoint
            checkpoint_filename = os.path.join(config.output_dir, f"checkpoint_epoch_{epoch+1}.pt")
            save_checkpoint(model, tokenizer, optimizer, scheduler, epoch, val_loss, val_metrics, config, checkpoint_filename)
            
            # Check if this is the best model
            if val_loss < best_val_loss:
                best_val_loss = val_loss
                early_stopping_counter = 0
                
                # Save best model
                best_model_filename = os.path.join(config.output_dir, "best_model.pt")
                save_checkpoint(model, tokenizer, optimizer, scheduler, epoch, val_loss, val_metrics, config, best_model_filename)
                logger.info(f"New best model saved with validation loss: {val_loss:.4f}")
            else:
                early_stopping_counter += 1
                logger.info(f"No improvement in validation loss for {early_stopping_counter} epochs")
                
                if early_stopping_counter >= config.early_stopping_patience:
                    logger.info(f"Early stopping triggered after {epoch+1} epochs")
                    break
            
            # Clean up GPU memory
            if torch.cuda.is_available():
                torch.cuda.empty_cache()
            
            # Visualize training history
            if len(history['train_loss']) > 0:
                visualize_training_history(history)
    
    except KeyboardInterrupt:
        logger.info("Training interrupted by user")
    
    # Load best model for final evaluation
    best_model_filename = os.path.join(config.output_dir, "best_model.pt")
    if os.path.exists(best_model_filename):
        model, optimizer, scheduler, best_epoch, best_val_loss, best_metrics = load_checkpoint(
            model, optimizer, scheduler, best_model_filename
        )
        logger.info(f"Loaded best model from epoch {best_epoch+1} with validation loss: {best_val_loss:.4f}")
    
    # Final evaluation on test set
    logger.info("Evaluating model on test set...")
    test_loss, test_metrics, predictions, references = evaluate(model, test_loader, config, tokenizer, compute_metrics=True)
    
    # Log test metrics
    log_str = "Test Results: "
    log_str += f"Loss: {test_loss:.4f}, "
    log_str += f"ROUGE-1 F1: {test_metrics.get('rouge1_f1', 0.0):.4f}, "
    log_str += f"ROUGE-2 F1: {test_metrics.get('rouge2_f1', 0.0):.4f}, "
    log_str += f"ROUGE-L F1: {test_metrics.get('rougeL_f1', 0.0):.4f}, "
    log_str += f"BLEU: {test_metrics.get('bleu', 0.0):.4f}"
    logger.info(log_str)
    
    # Save test results
    test_results = {
        'test_loss': test_loss,
        'test_metrics': test_metrics
    }
    torch.save(test_results, os.path.join(config.output_dir, "test_results.pt"))
    
    # Analyze examples
    logger.info("Analyzing examples...")
    examples = analyze_examples(model, test_loader, tokenizer, config, num_examples=5)
    
    for i, example in enumerate(examples):
        logger.info(f"Example {i+1}:")
        logger.info(f"Original Text: {example['original_text'][:300]}...")
        logger.info(f"Original Summary: {example['original_summary']}")
        logger.info(f"Generated Summary: {example['generated_summary']}")
        logger.info("-" * 50)
    
    return history, test_metrics, examples

# Prediction function
def predict_summary(model, tokenizer, text, config, device):
    """Generate summary for a given text"""
    # Preprocess
    text = preprocess_text(text)
    
    # Add prefix
    text_with_prefix = "summarize: " + text
    
    # Tokenize
    inputs = tokenizer(
        text_with_prefix,
        max_length=config.input_max_len,
        padding="max_length",
        truncation=True,
        return_tensors="pt"
    ).to(device)
    
    # Generate
    model.eval()
    with torch.no_grad():
        outputs = model.generate(
            input_ids=inputs["input_ids"],
            attention_mask=inputs["attention_mask"],
            max_length=config.output_max_len,
            num_beams=4,
            repetition_penalty=2.5,
            length_penalty=1.0,
            early_stopping=True
        )
    
    # Decode
    summary = tokenizer.decode(outputs[0], skip_special_tokens=True)
    
    return summary

# Main function
def main():
    """Main function"""
    # Configuration
    config = Config()
    
    # Check if checkpoint exists
    if os.path.exists(os.path.join(config.output_dir, "best_model.pt")):
        logger.info("Found existing checkpoint. Do you want to continue training? (y/n)")
        answer = input().strip().lower()
        if answer != 'y':
            logger.info("Starting training from scratch...")
            # Rename existing output directory
            os.rename(config.output_dir, f"{config.output_dir}_old_{int(time.time())}")
            os.makedirs(config.output_dir)
    
    # Train and evaluate
    history, test_metrics, examples = train_and_evaluate(config)
    
    # Visualize training history
    if history and len(history['train_loss']) > 0:
        visualize_training_history(history)
    
    logger.info("Training and evaluation completed!")
    
    # Example inference
    logger.info("\nExample Inference:")
    sample_text = "PT Bank Mandiri (Persero) Tbk. mencatat kenaikan laba bersih sebesar 45,2 persen secara tahunan pada kuartal pertama 2023 yaitu menjadi Rp 12,5 triliun. Direktur Utama Bank Mandiri, Darmawan Junaidi, mengatakan pertumbuhan laba tersebut didorong oleh pertumbuhan kredit sebesar 9,5 persen secara tahunan menjadi Rp 1.216,3 triliun. Selain itu, juga didorong oleh kenaikan dana pihak ketiga sebesar 6,2 persen secara tahunan menjadi Rp 1.162,6 triliun. 'Pertumbuhan kredit ini sejalan dengan pembiayaan Bank Mandiri yang tidak hanya fokus pada pemulihan ekonomi, tetapi juga mendukung pembangunan infrastruktur nasional,' kata Darmawan dalam keterangan tertulis yang dikutip Selasa (25/4/2023)."
    
    best_model_path = os.path.join(config.output_dir, "best_model.pt")
    if os.path.exists(best_model_path):
        # Load model, tokenizer and config
        checkpoint = torch.load(best_model_path, map_location=device)
        tokenizer = T5Tokenizer.from_pretrained(config.model_name)
        model = T5ForConditionalGeneration.from_pretrained(config.model_name)
        model.load_state_dict(checkpoint['model_state_dict'])
        model.to(device)
        
        # Generate summary
        summary = predict_summary(model, tokenizer, sample_text, config, device)
        logger.info(f"Text: {sample_text[:200]}...")
        logger.info(f"Summary: {summary}")

        # Display example analysis
        plt.figure(figsize=(12, 6))
        
        # Create a simple visualization of input vs output length
        input_len = len(sample_text)
        output_len = len(summary)
        reduction = (1 - output_len/input_len) * 100
        
        plt.bar(['Original Text', 'Summary'], [input_len, output_len], color=['#3498db', '#2ecc71'])
        plt.title(f'Text Length Comparison (Reduction: {reduction:.1f}%)')
        plt.ylabel('Number of Characters')
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        
        # Add values on top of bars
        plt.text(0, input_len + 10, f"{input_len} chars", ha='center')
        plt.text(1, output_len + 10, f"{output_len} chars", ha='center')
        
        plt.tight_layout()
        plt.savefig('summary_comparison.png')
        logger.info("Summary comparison visualization saved to 'summary_comparison.png'")
        
        try:
            from IPython import get_ipython
            if get_ipython() is not None:
                plt.show()
        except:
            pass
        
        plt.close()

if __name__ == "__main__":
    main()
```
